# Combine and filter OCO-2 Lite files for SIF and XCO2 datasets

NOTE: OCO-2 v10r Lite files are copied on the NIASRA HPC (https://hpc.niasra.uow.edu.au/); update paths as needed

```python
import sys

sys.path.insert(0, "../source")

import xarray as xr
from data_utils import prep_sif, prep_xco2
```

## SIF

```python
paths_sif = "/data/OCO-2/OCO2_L2_Lite_SIF.10r/*/*.nc4"
ds_sif = xr.open_mfdataset(paths_sif, preprocess=prep_sif, parallel=True)
ds_sif
```

```python
ds_sif.to_netcdf("../data/production/OCO2_Lite_SIF740.nc4", format="NETCDF4")
```

```python
ds_sif.close()
```

## XCO2

```python
paths_xco2 = "/data/OCO-2/OCO2_L2_Lite_FP.10r/*/*.nc4"
ds_xco2 = xr.open_mfdataset(paths_xco2, preprocess=prep_xco2, parallel=True)
ds_xco2
```

```python
ds_xco2.to_netcdf("../data/production/OCO2_Lite_XCO2.nc4", format="NETCDF4")
```

```python
ds_xco2.close()
```
