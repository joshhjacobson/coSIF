# Collect fitted parameters for bivariate models in all months in a single table for supplementary material

```python
import pandas as pd
```

```python
months = ["202102", "202104", "202107", "202110"]

models = []
for m in months:
    df_params = pd.read_csv(f"../data/intermediate/models/{m}/bivariate_model.csv", index_col=0)
    df_params["month"] = m
    models.append(df_params)

df = (
    pd.concat(models, axis=0)
    .reset_index(drop=True)
    .rename(columns={"name": "parameter"})[["month", "parameter", "value", "bounds"]]
)
df
```

```python
df.to_csv("../data/output/coSIF_fitted_model_parameters.csv")
```
